```python
import tensorflow as tf

# Create (створення скалярного тензора)
scalar = tf.constant(5)
print("Скаляр:", scalar)

# Read (читання значення)
print("Значення скаляра:", scalar.numpy())

# Update (оновлення значення скаляра)
updated_scalar = tf.constant(10)
print("Оновлений скаляр:", updated_scalar)

# Delete (видалення скаляра)
print(scalar)
del scalar
print("Скаляр був видалений")
```

2.Згенерувати набір даних для регресії та класифікації з використанням TensorFlow

```python
from sklearn.datasets import make_classification

# Генерація набору даних
X_classification, y_classification = make_classification(
    n_samples=200,  # кількість прикладів
    n_features=2,   # кількість ознак (вимірів)
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=42
)

# Конвертуємо в TensorFlow тензори
X_classification_tensor = tf.constant(X_classification, dtype=tf.float32)
y_classification_tensor = tf.constant(y_classification, dtype=tf.int32)

print("Вхідні дані для класифікації (X):", X_classification_tensor[:10])
print("Цільові мітки для класифікації (y):", y_classification_tensor[:10])
```

3.Здійснити графічний аналіз наборів даних за допомогою Matplotlib

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.datasets import make_classification, make_regression
from sklearn.linear_model import LinearRegression
import numpy as np

# Створення вектора та його оновленого варіанту
vector = tf.constant([1, 2, 3, 4, 5])
updated_vector = tf.constant([10, 20, 30, 40, 50])

# Графік для вектора до і після оновлення
plt.figure(figsize=(16, 4))
plt.subplot(1, 4, 1)
plt.plot(vector.numpy(), label='Original Vector', marker='o')
plt.plot(updated_vector.numpy(), label='Updated Vector', marker='x')
plt.title('Vector Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Створення матриці та її оновленого варіанту
matrix = tf.constant([[1, 2], [3, 4], [5, 6]])
updated_matrix = tf.constant([[10, 20], [30, 40], [50, 60]])

# Графік для матриці
plt.subplot(1, 4, 2)
plt.imshow(matrix.numpy(), cmap='viridis', aspect='auto')
plt.colorbar()
plt.title('Matrix Data')

# Генерація даних для класифікації
X_classification, y_classification = make_classification(
    n_samples=200,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=42
)

# Графік для класифікаційних даних
plt.subplot(1, 4, 3)
plt.scatter(X_classification[:, 0], X_classification[:, 1], c=y_classification, cmap='coolwarm', edgecolor='k')
plt.title('Classification Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Генерація даних для регресії
X_regression, y_regression = make_regression(
    n_samples=100, n_features=1, noise=10, random_state=42
)

# Створення та навчання моделі лінійної регресії
model = LinearRegression()
model.fit(X_regression, y_regression)
y_pred = model.predict(X_regression)

# Графік для регресійних даних з лінією регресії
plt.subplot(1, 4, 4)
plt.scatter(X_regression, y_regression, color='blue', label='Data points')
plt.plot(X_regression, y_pred, color='red', linewidth=2, label='Regression line')
plt.title('Regression Data')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()

plt.tight_layout()
plt.show()
```
