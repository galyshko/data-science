```python
import numpy as np

# Параметри
n_samples = 100  # Кількість зразків
slope = 2.5      # Коефіцієнт нахилу лінії
intercept = 5    # Зсув
noise_level = 0.5  # Рівень шуму

# Генерація вхідних даних
X = np.random.rand(n_samples, 1) * 10  # Незалежна змінна від 0 до 10

# Цільова змінна з додаванням випадкового шуму
y = slope * X + intercept + np.random.randn(n_samples, 1) * noise_level

# Перетворимо X та y у зрозумілий формат (наприклад, DataFrame) для зручного використання
import pandas as pd
data = pd.DataFrame(data={'X': X.flatten(), 'y': y.flatten()})

data.head()
```

 Регресія для першого набору даних (weather_classification_data.csv):


```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Завантаження даних
data = pd.read_csv('weather_classification_data.csv')

# Розділення даних на ознаки (X) та цільову змінну (y)
X_reg = data[['Temperature']]  # Незалежна змінна
y_reg = data['Humidity']        # Залежна змінна (можете змінити на іншу змінну, якщо потрібно)

# Створення моделі лінійної регресії
model = LinearRegression()

# Навчання моделі
model.fit(X_reg, y_reg)

# Передбачення на основі навчання
y_pred = model.predict(X_reg)

# Обчислення метрик
mse = mean_squared_error(y_reg, y_pred)
mae = mean_absolute_error(y_reg, y_pred)
r2 = r2_score(y_reg, y_pred)

# Виведення коефіцієнтів моделі та метрик
print(f"Коефіцієнт нахилу: {model.coef_[0]}")
print(f"Зсув: {model.intercept_}")
print(f"MSE: {mse}")
print(f"MAE: {mae}")
print(f"R²: {r2}")

# Візуалізація результатів
plt.scatter(X_reg, y_reg, color='blue', label='Дані')
plt.plot(X_reg, y_pred, color='red', label='Прогноз')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Лінійна регресія: Температура vs Вологість')
plt.legend()
plt.show()
```

Лінійна регресія для другого набору даних (згенерований у 3 завданні):

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Розділення даних на ознаки (X) та цільову змінну (y)
X_reg = data[['X']]  # Незалежна змінна
y_reg = data['y']    # Залежна змінна

# Створення моделі лінійної регресії
model = LinearRegression()

# Навчання моделі
model.fit(X_reg, y_reg)

# Передбачення на основі навчання
y_pred = model.predict(X_reg)


# Обчислення метрик
mae = mean_absolute_error(y_reg, y_pred)
mse = mean_squared_error(y_reg, y_pred)
r2 = r2_score(y_reg, y_pred)

# Виведення метрик
print(f"Коефіцієнт нахилу: {model.coef_[0]}")
print(f"Зсув: {model.intercept_}")
print(f"MSE: {mse}")
print(f"MAE: {mae}")
print(f"R²: {r2}")

# Візуалізація результатів
plt.scatter(X_reg, y_reg, color='blue', label='Дані')
plt.plot(X_reg, y_pred, color='red', label='Прогноз')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Лінійна регресія')
plt.legend()
plt.show()
```
